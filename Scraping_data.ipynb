```python
!pip install google-play-scraper

# Mengimpor pustaka google-play scrapper untuk akses ulasan dan infrmasi aplikasi
from google_play_scraper import app, Sort, reviews_all

import pandas as pd
import numpy as np
```

## **Scraping Dataset**

```python
scrapereview = reviews_all(
    'com.duolingo',  # ID aplikasi
    lang='id',     # bahasa ulasan
    country='id',
    sort=Sort.MOST_RELEVANT,  # Urutan ulasan default sort.most_relevant
    count=10000     # jumlah maksimum nulasan yang diambil
)
```

## **Save to CSV**

```python
duo = pd.DataFrame(scrapereview)
duo.to_csv('duo_apk_data.csv', index=False)
```
